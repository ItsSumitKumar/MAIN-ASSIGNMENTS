1. Explain what inheritance is in object-oriented programming and why it is used.

Solution: It is a mechanism that allows you to create a hierarchy of classes that share a set of properties and methods by deriving a class from another class.

Inheritance is the capability of one class (usually child class) to derive or inherit the properties from another class (usually parent class).

It is used due to following reason:

1. It represents real-world relationships well.
2. It provides the reusability of a code. We don’t have to write the same code again and again. Also, it allows us to add more features to a class without modifying it.
3. It is transitive in nature, which means that if class B inherits from another class A, then all the subclasses of B would automatically inherit from class A.
4. Inheritance offers a simple, understandable model structure.
5. Less development and maintenance expenses result from an inheritance.

2. Discuss the concept of single inheritance and multiple inheritance, highlighting their differences and advantages.

Solution: In single inheritance a child class has only one parent class while in multiple inheritance a child class has more than one parent class.

Differences:
1. Single inheritance is a mechanism in which a derived class inherits from a single base class whereas Multiple inheritance is a mechanism in which a derived class inherits from multiple base classes.
2. Single inheritance is simpler to implement whereas Multiple inheritance is more complex to implement
3. Single inheritance is less flexible whereas Multiple inheritance is more flexible.
4. Single inheritance is less error-prone. whereas Multiple inheritance is more error-prone.

Advantages of Single Inheritance:
1. It is simpler to implement.
2. It is less error-prone.
3. It is less complex.

Advantages of Multiple Inheritance:
1. It is more flexible.
2. It can inherit the properties of multiple classes.
3. It can be used to model more complex relationships.


```python
#Single Inheritance

class person:
    
    def __init__(self, name):
        self.name = name
        
    def display_info(self):
        print(f"Name is: {self.name}")

class student(person):
    
    def __init__(self, name, grade):
        # super keyword is used for inheriting paramteres from the derived parents class
        super().__init__(name)       
        self.grade = grade
        
    def display_info(self):
        super().display_info()
        return f"Grade is: {self.grade}"
    
Student1 = student("Sumit Kumar", "A+")
print(Student1.display_info())

print()

#Multiple Inheritance

class Father:
    
    def father(self, fathername):
        print(self.fathername)

class Mother:
    
    def mother(self, mothername):
        print(self.mothername)

class Child(Father, Mother):
    
    def parent(self):
        print("Father :", self.fathername)
        print("Mother :", self.mothername)

s1 = Child()
s1.fathername = "Satish Kumar"
s1.mothername = "Saweta"
s1.parent()
```

4. What is the significance of the "protected" access modifier in inheritance? How does it differ from "private" and "public" modifiers?

Solution: 
=> Protected Access Modifiers
1)The members of a class that are declared protected are only accessible to a class derived from it.
2) Data members of a class are declared protected by adding a single underscore ‘_’ symbol before the data member of that
class.

=> Public Access Modifier 1) The members of a class that are declared public are easily accessible from any part of the program. All data members and member functions of a class are public by default.

=> Private access Modifiers 1) The members of a class that are declared private are accessible within the class only, private access modifier is the most secure access modifier. 2) Data members of a class are declared private by adding a double underscore ‘__’ symbol before the data member of that class.

```python
#Protect Access Modifiers

class Student:
  # protected attributes
    def __init__(self,name,roll_number):
        self.name=name
        self.roll_number=roll_number
        
    def _display_info(self):
        print(f"The name of the student is {self.name} with roll number {self.roll_number}")
        
class Genius(Student):
    def __init__(self,name,roll_number,grade):
        super().__init__(name,roll_number)
        self.grade=grade
        
    def _display_details(self):
        print(f"{self.name} with roll number {self.roll_number} scored {self.grade} grade .")
        
        
genius1=Genius("Sumit","123abc","A")
genius1._display_details()

print()

#Public Access Modifiers

class Person:
    def __init__(self,name,lastname):
        self.name = name
        self.lastname = lastname

    def display(self):
        print(f"The name is {self.name} and last name is {self.lastname}")
        
person1=Person("Sumit","Kumar")
person1.display()

print()

#Private Access Modifiers

class Student:
    # private attributes
    def __init__(self,name,roll):
        self.__name=name
        self.__roll=roll

    def __display(self):
        print(f"Name is {self.__name} and roll number is {self.__roll}")

    def display(self):
        self.__display()
        
class Genius(Student):
    def __init__(self,name,roll,subject):
        super().__init__(name,roll)
        self.subject=subject
        
genius=Genius("sumit","123abc","AI")
genius.display()
```

5. What is the purpose of the "super" keyword in inheritance? Provide an example.

Solution: The super() function is used to call the parent class constructor. It is used to call the constructor, methods, and properties of the parent class.



```python
class person:
    
    def __init__(self, name):
        self.name = name
        
    def display_info(self):
        print(f"Name is: {self.name}")
        
class student(person):
    
    def __init__(self, name, grade):
        # super keyword is used for inheriting paramteres from the derived parents class
        super().__init__(name)       
        self.grade = grade
        
    def display_info(self):
        super().display_info()
        return f"Grade is: {self.grade}"
    
Student1 = student("Sumit", "A+")
print(Student1.display_info())  
```

8. Design a base class called "Shape" with attributes like "colour" and "border_width." Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add specific attributes like "length" and "width" for the "Rectangle" class and "radius" for the "Circle" class.

Solution: 

```python
import math

class Shape:
    def __init__(self,colour,border_width):
        self.colour=colour
        self.border_width=border_width
        
    def perimeter(self):
        pass
    
    def area(self):
        pass
    
class Rectangle(Shape):
    def __init__(self,colour,border_width,length,width):
        super().__init__(colour,border_width)
        self.length=length
        self.width=width
        
    def perimeter(self):
        perimeter = 2 * ( self.length + self.width )
        print(f"The perimeter of rectangle is {perimeter}")
        
    def area(self):
        area = self.length * self.width
        print(f"The area of a rectangle is {area}")
        
class Circle(Shape):
    def __init__(self,colour,border_width,radius):
        super().__init__(colour,border_width)
        self.radius=radius
        
    def perimeter(self):
        perimeter = 2 * math.pi * self.radius
        print(f"The radius of circle is {perimeter}")
        
    def area(self):
        area = math.pi * self.radius * self.radius
        print(f"The area of circle is {area}")
        
rect1=Rectangle("blue",2,5,3)
rect1.perimeter()
rect1.area()

print("-"*30)

circ1=Circle("white",2,7)
circ1.perimeter()
circ1.area()
        
```

9. Create a base class called "Device" with attributes like "brand" and "model." Derive two classes, "Phone" and "Tablet," from "Device." Add specific attributes like "screen_size" for the "Phone" class and "battery_capacity" for the "Tablet" class.

Solution: 

```python
class Device:
    
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        
    def display_info(self):
        print(f"Brand: {self.brand}\nModel: {self.model}")

class Phone(Device):
    
    def __init__(self, brand, model, screen_size):
        super().__init__(brand, model)
        self.screen_size = screen_size
        
    def display_info(self):
        super().display_info()
        print(f"Screen Size: {self.screen_size} inches")
        
class Tablet(Device):
    
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity
        
    def display_info(self):
        super().display_info()
        print(f"Battery Capacity: {self.battery_capacity} mAh")
        
phone1=Phone("Apple","I Phone 13","5 inch")
phone1.display_info()

print("-"*30)

tab1=Tablet("Apple","I Pad 9","7040 mah")
tab1.display_info()
```

10. Create a base class called "BankAccount" with attributes like "account_number" and "balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from "BankAccount." Add specific methods like "calculate_interest" for the "SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.

Solution: 

```python
class BankAccount:
    
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
        
    def display_info(self):
        print(f"Account Number: {self.account_number}")
        
class SavingsAccount(BankAccount):
    
    def __init__(self, account_number, balance):
        super().__init__(account_number, balance)
        
    def calculate_interest(self, interest):
        interest_amount = (interest / 100) * self.balance
        self.balance += interest_amount
        
    def display_info(self):
        super().display_info()
        print(f"Available Balance: {self.balance}")
        
class CheckingAccount(BankAccount):
    
    def __init__(self, account_number, balance):
        super().__init__(account_number, balance)
        
    def deduct_fees(self, annual_fee_charged):
        self.balance -= annual_fee_charged
        
    def display_info(self):
        super().display_info()
        print(f"Available Balance: {self.balance}")
        
sav_acc1=SavingsAccount(65190067575, 105000)
print("Balance before interest:")
sav_acc1.display_info()

print()

sav_acc1.calculate_interest(3.5)
print("Balance after interest:")
sav_acc1.display_info()

print("-"*40)

check_acc1=CheckingAccount(50100376115153, 23600)
print("Balance before annual fees:")
check_acc1.display_info()

print()

check_acc1.deduct_fees(1500)
print("Balance after anuual fees:")
check_acc1.display_info()
```
